## Loading Data

Pandas has easy-to-use functions to load data. Let's load the Iris data, just for one type of flower: Iris setosa.

```
import pandas as pd

# Load data for Iris setosa.
setosa_df = pd.read_csv("https://raw.githubusercontent.com/rthorst/ECDS_Workshops/master/python_workshop/data/setosa.csv")

# View the first few rows.
setosa_df.head()
```

---

## Renaming Columns

Whoops! Those columns aren't named quite right. Let's give them better names.

![alt text](https://raw.githubusercontent.com/rthorst/ECDS_Workshops/master/python_workshop/images/pd_headers.png)

```
# Let's rename the columns in a more standard way.
new_colnames = ["sepal_length", "sepal_width", "petal_length", "petal_width", 
                "species"]
setosa_df.columns = new_colnames

# Now, visualize the top of the dataframe. 
setosa_df.head()
```

---

## Missing Values

It looks like there are some missing values in the data! (See row 2). By default, Pandas treats these as a special type of number called ... wait for it .... "Not a Number (NaN)."

We often want to treat rows with missing data in a special way. **Let's remove rows with missing data for now**

![alt text](https://raw.githubusercontent.com/rthorst/ECDS_Workshops/master/python_workshop/images/pd_nans.png)

```
# Drop rows with at least one NaN value.
setosa_df = setosa_df.dropna() # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html

# View the top of the dataframe. Notice we've dropped row number 2.
setosa_df.head()
```

---

## Correcting Values

It also looks like we need to fix one of the values in the dataframe. Did you notice the third entry for sepal length? It's written out as "four point six" -- not good!

Let's correct it to a numerical value.

![alt text](https://raw.githubusercontent.com/rthorst/ECDS_Workshops/master/python_workshop/images/pd_wrong_val.png)

```
# Set the value at row index #3, column "sepal length" to 4.6
setosa_df.at[3, "sepal_length"] = 4.6

# We also need to change the type of every value in the "sepal_length" column 
# From "string" (e.g. text) to a floating point number. 
setosa_df["sepal_length"] = setosa_df["sepal_length"].astype("float64")

# Let's view the top of the dataframe again.
setosa_df.head()
```

---

## Merging Two Datasets Together

Ok, great! We have data for one type of Iris, Iris setosa.

We'd also like to analyze data for another type of Iris, Iris virginica. But, this data is in a different file. How can we join them together?

Pandas lets you do any type of table join you want (think SQL-type joins: https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html). Here, we just want to append one dataframe to the end of the other.

```
# Load data for Iris versicolor.
versicolor_df = pd.read_csv("https://raw.githubusercontent.com/rthorst/ECDS_Workshops/master/python_workshop/data/versicolor.csv")

# View the data for Iris versicolor.
versicolor_df.head()
```

```
# Append the versicolor data to the botton of the data for Iris virginica.
merged_df = setosa_df.append(versicolor_df)

# View the merged data.
merged_df
```

### Encoding a string

If you've ever seen a few tweets, you will know that many of them can have special characters. 

Usually, we want to remove those special characters.

A quick way to do this is to capitalize on two special functions for strings:

* **encode()** a string (as raw bytes) using a specified format.
* **decode()** decode the result back as a string.

A simple solution is to keep only characters in the **ASCII** character set. 

```
# Here is a string with special characters.
s = "Hello, world!" + "".join([chr(v) for v in range(200, 500, 10)])
print(s)
```

```
# A simple trick to remove the special characters, 
# encoding as ASCII. Note: only works in Python 3.

s = s.encode("ascii", errors="ignore").decode()
print(s)
```

## Working with Dates

In everyday life, we express dates rather loosely in natural language. For a computer, we want to know much more precisely about a date (down to the millisecond if possible). For this reason, there are specialized packages of code (called libraries) for working with dates in every modern language. We want to use these libraries to read in and process dates.

In python, this library is called **datetime**.

---

### Reading in dates with datetime

The first thing you want to do is convert a date to a datetime object.

You will get a date in one of two formats:
* As a **string**, e.g. "Mon Oct 07 2019 20:27:41"
* As a **unix timestamp**., e.g. 1570494461. This is expressed as a number of milliseconds since 1/1/1970.

Converting unix timestamps to dates is pretty easy:

```
from datetime import datetime 

# Converting unix timestamp to date.
timestamp = 1570494461
dt = datetime.fromtimestamp(timestamp)
print(dt)
```

Converting a string to a datetime is a little bit harder. You need to specify the format of the string using a specialized language:

Here are some examples of the format.

![alt text](https://raw.githubusercontent.com/rthorst/ECDS_Workshops/master/python_workshop/images/strftime_ref.PNG)

```
# Convert string to date.
str_date = "Mon Oct 07 2019 20:27:41"
fmt = "%a %b %d %Y %H:%M:%S"

dt = datetime.strptime(str_date, fmt)
print(dt)
```

Once you have a date, you may want to do a few things:

* **Compare two dates** which can be done with simple arithmetic operators like > and +
* **Extract information from a date** such as the month, day, and year. 

```
# Compare two dates: creation date of the tweet vs. now.

timestamp = 1570494461
dt_tweet = datetime.fromtimestamp(timestamp) # Oct 7 2019 20:27:41.
dt_now = datetime.now()

# Which one came first?
print(dt_tweet < dt_now)

# How much earlier was the tweet than today.
print(dt_now - dt_tweet)
```

```
# Extract information from a date such as day, month, year.
day = dt_tweet.day
month = dt_tweet.month
year = dt_tweet.year

print("Day : {}\nMonth : {}\nYear : {}".format(day, month, year))
```
